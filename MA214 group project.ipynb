```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics as st
```

```python
# Import Shanghai dataframe 
shanghai = pd.read_excel("/Users/khanht.dinh/Documents/MA214 group project/Datasets/shanghai2.xls")
G1 = nx.Graph()
G1.add_edges_from([(row["start_station"], row["destination"])for idx, row in shanghai.iterrows()])
```

```python
nx.draw(G1, with_labels = False, node_color = "magenta")
```

```python
print(G1.number_of_nodes())
```

```python
between_G1 = nx.betweenness_centrality(G1)
dict(sorted(between_G1.items(), key=lambda item: item[1], reverse=True))
```

```python
closeness_G1 = nx.closeness_centrality(G1)
dict(sorted(closeness_G1.items(),key=lambda item: item[1], reverse = True))
```

```python
# Eigenvector centrality
eigen_G1 = nx.eigenvector_centrality(G1)
dict(sorted(eigen_G1.items(), key=lambda item: item[1],reverse = True))
```

```python
G1.nodes()
```

```python
# Find the nodes with the highest degrees
sorted(G1.degree, key=lambda x: x[1], reverse=True)
```

```python
# Average shortest path length
print(nx.average_shortest_path_length(G1))
# diameter
print(nx.diameter(G1))
```

```python
# Import London dataframe
london = pd.read_csv("/Users/khanht.dinh/Documents/MA214 group project/Datasets/london.connections.txt", sep=",", header = None)
london.columns = ["station1", "station2", "line", "time"]
```

```python
G2 = nx.Graph()
G2.add_edges_from([(row["station1"], row["station2"]) for idx, row in london.iterrows()])
```

```python
nx.draw(G2, node_color = "yellow")
```

```python
#Closeness centrality
closeness_G2 = nx.closeness_centrality(G2)
dict(sorted(closeness_G2.items(), key=lambda item: item[1],reverse = True))
```
